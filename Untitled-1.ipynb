IMPORT LIBRARIES

```python
import os # for making directories and list the items in it
import shutil 
import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns

```

DIRECTORY TO STORE THE HISTOPATHOLOGICAL IMAGES

```python
cancer_rays_directory=os.listdir("WOMWN/IDC_regular_ps50_idx5")
cancer_rays_directory
len(cancer_rays_directory)
```

```python
all_rays_directory="all_rays_directory"
cancer_rays_path="WOMWN/IDC_regular_ps50_idx5/"
```

```python
os.mkdir(all_rays_directory)
```

```python
all_rays_directory_items=os.listdir('./all_rays_directory')
```

SEPARATE THE FILES TO TARGET ZERO AND TARGET ONE

```python

for patient in cancer_rays_directory:
    zero_path=cancer_rays_path+str(patient)+'/0'
    one_path=cancer_rays_path+str(patient)+'/1'
    file_zero=os.listdir(zero_path)
    file_one=os.listdir(one_path)
    for file_name in file_zero:
        source = os.path.join(zero_path,file_name)
        destination = os.path.join(all_rays_directory,file_name)
        shutil.copyfile(source, destination)
    for file_name in file_one:
        source = os.path.join(one_path,file_name)
        destination = os.path.join(all_rays_directory,file_name)
        shutil.copyfile(source, destination)
    
```

```python
print(file_zero)
print(file_one)
all_rays_directory_items = os.listdir('./all_rays_directory')
len(all_rays_directory_items)
```

```python
data = pd.DataFrame(all_rays_directory_items, columns = ['id_image'])
data.head()
```

```python
#split the target into a separate column
def split_target(f):
    x = f.split('_')
    y = x[4]
    target_no = y[5]
    return target_no

data['target'] = data['id_image'].apply(split_target)
data.head(5)
```

```python
def split_patient_id(f):
    x=f.split('_')
    patient_id=x[0]
    return patient_id
data['patient_id']=data['id_image'].apply(split_patient_id)
data.head(5)
```

EXPLORATORY DATA ANALYSIS (EDA)

HEALTHY PATCHES

```python
data.target = data.target.astype(np.int)
fig,ax = plt.subplots(6,10,figsize = (20,10))
positive_cells  = np.random.choice(data[data.target ==1].index,size=60, replace=False)
negative_cells = np.random.choice(data[data.target ==0].index,size=60, replace=False)

for i in range(6):
    for j in range(10):
        index = negative_cells[j+i*10]
        path = os.path.join(all_rays_directory,data.loc[index,'id_image'])

        img = mpimg.imread(path)
        ax[i,j].imshow(img)
        ax[i,j].grid(False)
```

NEGATIVE CELLS

```python
fig,ax = plt.subplots(6,10,figsize = (20,10))
positive_cells  = np.random.choice(data[data.target ==1].index,size=60, replace=False)
negative_cells = np.random.choice(data[data.target ==0].index,size=60, replace=False)

for i in range(6):
    for j in range(10):
        index = positive_cells[j+i*10]
        path = os.path.join(all_rays_directory,data.loc[index,'id_image'])

        img = mpimg.imread(path)
        ax[i,j].imshow(img)
        ax[i,j].grid(False)
```

```python
cancer_precentage=data.groupby("patient_id").target.value_counts()/data.groupby("patient_id").target.size()
cancer_precentage=cancer_precentage.unstack()

print(cancer_precentage)
```

```python
sns.distplot(data.groupby("patient_id").size(),color='Orange',kde=False,bins=30)
plt.xlabel('Number of patches')
plt.ylabel('Frequency')
plt.title("How many patches we have per patient?")
```

```python
sns.distplot(cancer_precentage.loc[:,1]*100,color="Tomato",kde=False,bins=30)
plt.xlabel('How much percentage of an image is covered by IDC?')
plt.ylabel('Frequency')
plt.title("% of patches with IDC")
```
